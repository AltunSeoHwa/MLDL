# Column chart / Bar chart

```python
# pandas 를 불러옵니다.
import pandas as pd

# marathon_2015_2017.csv 파일을 읽어옵니다.
marathon_2015_2017 = pd.read_csv("../marathon_data/marathon_2015_2017.csv")

# matplotlib 를 불러옵니다.
import matplotlib.pyplot as plt

# seaborn 을 불어옵니다.
import seaborn as sns
```

```python
# marathon_2015_2017 DataFrame 상위 5개 데이터를 확인합니다.
marathon_2015_2017.head()
```

```python
# marathon_2015_2017 DataFrame에서 Country == 'USA' 인 데이터만 가져옵니다.
USA_runner = marathon_2015_2017[marathon_2015_2017.Country == 'USA']
```

```python
# USA_runner DataFrame 의 기본적인 정보를 확인합니다. 
USA_runner.info()
```

```python
# 그림틀의 크기를 지정한다.
plt.figure(figsize=(20,10))
```

```python
# 그림틀의 크기를 지정한다.
plt.figure(figsize=(30,20))

# USA_runner DataFrame 에서 'State' 별로 참가자가 몇명인지를 그래프로 확인.
runner_state = sns.countplot(x='State',data=USA_runner)   # 차트에 나타낼 데이터 선택, x축에 'State' 지정, y축에는 counting
runner_state.set_title('Number of Runner by State - USA', fontsize=18)   # 차트 title 추가
runner_state.set_xlabel('State', fontdict= {'size':16})                  # x축 lable 추가
runner_state.set_ylabel('Number of Runner', fontdict= {'size':16})       # y축 lable 추가
plt.show()  # 그래프를 출력
```

```python
# 그림틀의 크기를 지정한다.
plt.figure(figsize=(30,20))

# USA_runner DataFrame 에서 'State' 별로 참가자가 몇명인지를 그래프로 확인.
runner_state = sns.countplot(y='State',data=USA_runner)   # 차트에 나타낼 데이터 선택, y축에 'State' 지정, x축에는 counting
runner_state.set_title('Number of Runner by State - USA', fontsize=18)   # 차트 title 추가
runner_state.set_ylabel('State', fontdict= {'size':16})                  # y축 lable 추가
runner_state.set_xlabel('Number of Runner', fontdict= {'size':16})       # x축 lable 추가
plt.show()  # 그래프를 출력
```

```python
# 그림틀의 크기를 지정한다.
plt.figure(figsize=(30,20))
# USA_runner DataFrame 에서 'State', 'M/F' 별로 참가자가 몇명인지를 그래프로 확인.
# 차트에 나타낼 데이터 선택, x축에 'State' 지정, hue 에 컬럼을 지정하면 그 컬럼을 기준으로 데이터를 쪼갠다.
runner_state = sns.countplot(x='State',data=USA_runner, hue='M/F',  palette={'F':'r','M':'b'})
runner_state.set_title('Number of Runner by State, Gender - USA', fontsize=18)  # 차트 title 추가
runner_state.set_xlabel('State', fontdict= {'size':16})                         # x축 lable 추가
runner_state.set_ylabel('Number of Runner', fontdict= {'size':16})              # y축 lable 추가
plt.show()  # 그래프를 출력
```

```python
# 그림틀의 크기를 지정한다.
plt.figure(figsize=(30,20))
# USA_runner DataFrame 에서 'State', 'Year' 별로 참가자가 몇명인지를 그래프로 확인.
# 차트에 나타낼 데이터 선택, x축에 'State' 지정, hue 에 컬럼을 지정하면 그 컬럼을 기준으로 데이터를 쪼갠다.
runner_state = sns.countplot(x='State',data=USA_runner, hue='Year')
runner_state.set_title('Number of Runner by State, year - USA', fontsize=18)  # 차트 title 추가 
runner_state.set_xlabel('State', fontdict= {'size':16})                       # x축 lable 추가
runner_state.set_ylabel('Number of Runner', fontdict= {'size':16})            # y축 lable 추가
plt.show()  # 그래프를 출력
```
