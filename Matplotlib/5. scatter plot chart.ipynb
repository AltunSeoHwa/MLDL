# Scatter plot chart

```python
# pandas 를 불러옵니다.
import pandas as pd

# marathon_2015_2017.csv 파일을 읽어옵니다.
marathon_2015_2017 = pd.read_csv("../marathon_data/marathon_2015_2017.csv")

# matplotlib 를 불러옵니다.
import matplotlib.pyplot as plt
```

```python
# marathon_2015_2017 에서 성별이 남자인 데이터만 가져옵니다.
MALE_runner = marathon_2015_2017[marathon_2015_2017['M/F'] == 'M']
# marathon_2015_2017 에서 성별이 여자인 데이터만 가져옵니다.
FEMALE_runner = marathon_2015_2017[marathon_2015_2017['M/F'] == 'F']
```

```python
FEMALE_runner.head()
```

```python
# 그림틀의 크기를 지정한다.
plt.figure(figsize=(20,20))
# MALE_runner에서 'Age' column 만 가져와 x_male Series 를 생성합니다.(x값 생성)
x_male = MALE_runner.Age
# MALE_runner에서 'Official Time' column 만 가져와 y_male Series 를 생성합니다.(y값 생성)
y_male = MALE_runner['Official Time']
# FEMALE_runner에서 'Age' column 만 가져와 x_female Series 를 생성합니다.(x값 생성)
x_female = FEMALE_runner.Age
# FEMALE_runner에서 'Official Time' column 만 가져와 y_female Series 를 생성합니다.(y값 생성)
y_female = FEMALE_runner['Official Time']
```

```python
x_male
```

```python
y_male
```

```python
# 그림틀의 크기를 지정한다.
plt.figure(figsize=(20,20))

# x축, y축 데이터를 지정하여 scatter plot chart를 그립니다.
plt.plot(x_male, y_male, '.b', alpha=0.5)     # linestyle='.', color='파랑', alpha=0.5 는 투명도 지정
plt.plot(x_female, y_female, '.r', alpha=0.5) # linestyle='.', color='빨강', alpha=0.5 는 투명도 지정

plt.xlabel("Age", fontsize=16)                                 # x축 lable 추가
plt.ylabel("Official Time (Second)",fontsize=16)               # y축 lable 추가
plt.title("Distribution by Running Time and Age",fontsize=20)  # 차트 title 추가

# 그래프를 출력
plt.show()
```
